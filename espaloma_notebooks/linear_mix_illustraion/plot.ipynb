```python
import torch
```

```python
def u(x, k, b):
    return 0.5 * k * (x - b) ** 2
```

```python
def u_linear_basis(x, k1, k2, b1, b2):
    return 0.5 * k1 * (x - b1) ** 2 + 0.5 * k2 * (x - b2) ** 2
```

```python
def u_linear_basis_absolute(x, k1, k2, b1, b2):
    k = k1 + k2
    b = (k1 * b1 + k2 * b2) / (k1 + k2)
    return u(x, k, b)
```

```python
k0 = 5; b0 = 5
```

```python
x = torch.distributions.Uniform(0, 10).sample([1000])
```

```python
import math
import numpy as np
b1 = 0.0; b2 = 10.0
k1 = torch.nn.Parameter(torch.distributions.Uniform(-10, math.log(10)).sample())
k2 = torch.nn.Parameter(torch.distributions.Uniform(-10, math.log(10)).sample())
optimizer = torch.optim.Adam([k1, k2], 1e-1)

traj1 = []
traj2 = []

for _ in range(1000):
    optimizer.zero_grad()
    loss = (u_linear_basis_absolute(x=x, k1=k1.exp(), k2=k2.exp(), b1=b1, b2=b2) - u(x, k=k0, b=b0)).pow(2).mean()
    loss.backward()
    optimizer.step()
    traj1.append(k1.exp().item())
    traj2.append(k2.exp().item())

traj1 = np.array(traj1)
traj2 = np.array(traj2)
```
