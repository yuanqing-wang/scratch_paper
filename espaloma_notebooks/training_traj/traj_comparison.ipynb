```python
import numpy as np
```

```python
from matplotlib import pyplot as plt
x = np.load("long.npy")[:2100]
plt.plot(np.linspace(0, 20000, len(x)), x)
plt.xlabel("epoch")
plt.ylabel("RMSE (centerd, kcal/mol)")
plt.title("without any treatment, min RMSE=%.2f kcal/mol" % x.min())
plt.ylim(0, 5)
plt.savefig("baseline.png", dpi=100)
```

```python
from matplotlib import pyplot as plt
x = np.load("force_only.npy")[:2100]
plt.plot(np.linspace(0, 20000, len(x)), x)
plt.xlabel("epoch")
plt.ylabel("RMSE (centerd, kcal/mol)")
plt.title("fitting force only, min RMSE=%.2f kcal/mol" % x.min())
plt.ylim(0, 5)
plt.savefig("force_only.png", dpi=100)
```

```python
from matplotlib import pyplot as plt
for weight_decay in ["1e-5", "1e-6", "1e-7", "1e-8", "1e-9"]:
    x = np.load("weight_decay%srmse.npy" % weight_decay)[100:2100]
    plt.plot(
        np.linspace(0, 20000, len(x)), x,
        label="weight_decay=%s, min RMSE=%.2f kcal/mol" % (weight_decay, x.min())
    )
    plt.xlabel("epoch")
    plt.ylabel("RMSE (centerd, kcal/mol)")
plt.legend()
plt.ylim(0, 5)
plt.savefig("weight_decay.png", dpi=100)
```

```python
from matplotlib import pyplot as plt
import numpy as np
for repeat in range(0, 5):
    x = np.load("%srmse.npy" % repeat)
    plt.plot(
        np.linspace(0, 20000, len(x)), x,
        alpha=0.5,
        label="min RMSE=%.4f" % x.min()
    )
    plt.xlabel("epoch")
    plt.ylabel("RMSE (centerd, kcal/mol)")
plt.legend()
plt.ylim(0, 5)
```
