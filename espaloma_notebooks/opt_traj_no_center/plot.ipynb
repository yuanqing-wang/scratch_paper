```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
from matplotlib import rcParams
rcParams['font.family'] = 'sans-serif'
rcParams['font.sans-serif'] = ['Helvetica']
plt.figure(figsize=(4,3 ))

for idx in range(5):
    for part in ['tr']:
        x = np.load(
                "%s_%s_0.npy" % (idx, part)
        )[::10]
        
        plt.plot(
            np.arange(0, 10000, 10),
            x,
            alpha=0.6,
            label="repetition %s" % idx,
        )
plt.yscale("log")
plt.xlabel("epoch")
plt.ylabel("training RMSE (kcal/mol)")
plt.legend()
plt.tight_layout()
plt.savefig("train_traj_no_center.png", dpi=300)
```

```python
for idx in range(5):
    rmse_vl = np.load("%s_%s_1.npy" % (idx, "vl"))
    rmse_tr = np.load("%s_%s_1.npy" % (idx, "tr"))
    rmse_te = np.load("%s_%s_1.npy" % (idx, "te"))
    print(
        rmse_vl.argmin(),
        "%.4f" % rmse_tr[rmse_vl.argmin()],
        "%.4f" % rmse_vl[rmse_vl.argmin()],
        "%.4f" % rmse_te[rmse_vl.argmin()],
    )
```
