```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
rmsd_esp = np.load("rmsd/rmsd_esp.npy")
rmsd_gaff1 = np.load("rmsd/rmsd_gaff-1.81.npy")
rmsd_gaff2 = np.load("rmsd/rmsd_gaff-2.11.npy")
rmsd_openff = np.load("rmsd/rmsd_openff-1.2.0.npy")
```

```python
from matplotlib import rcParams
rcParams['font.family'] = 'sans-serif'
rcParams['font.sans-serif'] = ['Helvetica']
plt.figure(figsize=(4, 2))
plt.plot(rmsd_esp, label="espaloma")
plt.plot(rmsd_gaff1, label="gaff-1.81")
plt.plot(rmsd_gaff2, label="gff-2.11")
plt.plot(rmsd_openff, label="openff-1.2.0")
plt.legend()
plt.xlabel("time (picosecond)")
plt.ylabel("RMSD (Å)")
plt.tight_layout()
plt.savefig("rmsd.png", dpi=300)
```
