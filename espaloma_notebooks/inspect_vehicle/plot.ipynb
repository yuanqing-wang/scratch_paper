```python
import numpy as np
```

```python
u_qm = np.load("u_qm.npy")
u_esp = np.load("u_esp.npy")
u_off = np.load("u_off.npy")
u_gaff1 = np.load("u_gaff1.npy")
u_gaff2 = np.load("u_gaff2.npy")
```

```python
u_qm -= u_qm.mean()
u_esp -= u_esp.mean()
u_off -= u_off.mean()
u_gaff1 -= u_gaff1.mean()
u_gaff2 -= u_gaff2.mean()
```

```python
u_qm *= 627.5
u_esp *= 627.5
u_off *= 627.5
u_gaff1 *= 627.5
u_gaff2 *= 627.5
```

```python
rmse_esp = ((u_esp - u_qm) ** 2).mean() ** 0.5
rmse_off = ((u_off - u_qm) ** 2).mean() ** 0.5
rmse_gaff1 = ((u_gaff1 - u_qm) ** 2).mean() ** 0.5
rmse_gaff2 = ((u_gaff2 - u_qm) ** 2).mean() ** 0.5
```

```python
from matplotlib import pyplot as plt
```

```python
plt.rcParams["font.family"] = "sans-serif"
plt.rcParams["font.sans-serif"] = "Helvetica"
plt.figure(figsize=(5, 3))
plt.plot(u_esp, "-o", label="espaloma" % rmse_esp, zorder=100)
plt.plot(u_qm, "-o", label="qm", zorder=88)
plt.plot(u_off, "-o", label="openff-1.2.0" % rmse_off, alpha=0.4)
plt.plot(u_gaff1, "-o", label="gaff-1.81" % rmse_gaff1, alpha=0.4)
plt.plot(u_gaff2, "-o", label="gaff-2.11" % rmse_gaff2, alpha=0.4)
plt.xlabel("Iterations")
plt.ylabel("$U$ (kcal/mol)")
plt.legend(loc="upper right")
plt.tight_layout()
# plt.title("Espaloma RMSE=%.2f kcal/mol; OpenFF-1.2.0 RMSE=%.2f kcal.mol" % (rmse_esp, rmse_off)) 
plt.savefig("opt.png", dpi=300)
```
